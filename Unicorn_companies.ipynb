```python
companies_sampled = df.sample(n=50, random_state=42)
companies_sampled.head()
```

```python
companies_sampled['time needed'] = companies_sampled['Year']- companies_sampled["Year Founded"]
companies_sampled.head()
```

```python
grouped = (companies_sampled[["Industry", "time needed"]]
           .groupby("Industry")
           .max()
           .sort_values(by="time needed")
           )
grouped
```

```python
plt.bar(grouped.index,
        height=grouped['time needed'], 
        label='Years Needed')


plt.title('Years Needed To Join Unicorn Companies')

plt.xlabel("company name")
plt.ylabel("years")


plt.xticks(rotation=90)


plt.legend()
plt.show()
```

```python
df["Month Joined"] = df["Date Joined"].dt.month_name()
df.head()
```

```python
df["Date Joined"] = pd.to_datetime(df['Date Joined'])
df.dtypes
```

```python
df["Years to Join"] = df["Date Joined"].dt.year - df["Year Founded"]
df.head()
```

```python
y2021 = df[df["Date Joined"].dt.year== 2021 ]
y2021.head()
```

```python
y2021.insert(
    3, "Week Joined", y2021["Date Joined"].dt.strftime('%Y-W%V'), True)
```

```python
w2021 = y2021.groupby(by="Week Joined")["Company"].count().reset_index().rename(columns={"Company":"Company Count"})
```

```python
w2021.value_counts
```

```python
y2020 = df[df["Date Joined"].dt.year == 2020]

y2020_1 = pd.concat(
    [y2020, y2021.drop(columns="Week Joined")])

y2020_1["Quarter Joined"] = y2020_1["Date Joined"].dt.to_period(
    'Q').dt.strftime('%Y-Q%q')

y2020_1["Valuation"] =  y2020_1["Valuation"].str.strip("$B").astype(float)

companies_by_quarter_2020_2021 = y2020_1.groupby(by="Quarter Joined")["Valuation"].mean().reset_index().rename(columns={"Valuation":"Average Valuation"})

companies_by_quarter_2020_2021.head()
```

```python
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November",
               "December"]
```

```python
sns.boxplot(x=df['Month Joined'],
            y=df['Years to Join'],
            order=month_order,
            showfliers=False)
plt.title('Distribution of years to become unicorn with respect to month joined')
plt.xticks(rotation=45, horizontalalignment='right')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df["Year Founded"],
            y=df["Years to Join"], errorbar=('ci', False))
plt.title("Bar plot of years to join with respect to year founded")
plt.xlabel("Year founded")
plt.ylabel("Years to join unicorn status")
plt.xticks(rotation=45, horizontalalignment='right')
plt.show()
```

```python
plt.figure(figsize=(20, 5))


plt.bar(x=w2021['Week Joined'],
        height=w2021['Company Count'])
plt.plot()


plt.xlabel("Week number")

plt.ylabel("Number of companies")


plt.title("Number of companies that became unicorns per week in 2021")


plt.xticks(rotation=45, horizontalalignment='right', fontsize=8)


plt.show()
```

```python
companies_by_quarter_2020_2021['Quarter Number'] = companies_by_quarter_2020_2021['Quarter Joined'].str[-2:]
companies_by_quarter_2020_2021['Year Joined'] = companies_by_quarter_2020_2021['Quarter Joined'].str[:4]


plt.figure(figsize=(10, 5))


sns.barplot(x=companies_by_quarter_2020_2021['Quarter Number'],
            y=companies_by_quarter_2020_2021['Average Valuation'],
            hue=companies_by_quarter_2020_2021['Year Joined'])
plt.plot()


plt.xlabel("Quarter number")


plt.ylabel("Average valuation (billions of dollars)")


plt.title(
    "Average valuation of companies that became unicorns per quarter in 2020 vs. 2021")

plt.show()
```
